```python
import csv
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import seaborn as sns
import matplotlib.pyplot as plt
import dateutil.parser as parser

from datetime import datetime

SPRING='SPINRG'
SUMMER='SUMMER'
AUTUMN='AUTUMN'
WINTER='WINTER'

SEASONS = [SPRING, SUMMER, AUTUMN, WINTER]
```

```python
tracks_df = pd.read_csv("./recent_tracks.csv")
tracks_df.columns
```

```python
spring_tracks = tracks_df.loc[tracks_df['season'] == SPRING]
summer_tracks = tracks_df.loc[tracks_df['season'] == SUMMER]
autumn_tracks = tracks_df.loc[tracks_df['season'] == AUTUMN]
winter_tracks = tracks_df.loc[tracks_df['season'] == WINTER]

season_track_map = {
    SPRING: spring_tracks,
    SUMMER: summer_tracks,
    AUTUMN: autumn_tracks,
    WINTER: winter_tracks,
}
```

```python
plt.boxplot([spring_tracks['danceability'], winter_tracks['danceability'], summer_tracks['danceability']])
```

```python
def compare_feature_between_seasons(feature):
    data = []
    labels = []
    for season in SEASONS:
        season_data = season_track_map[season][feature]
        data.append(season_data)
        labels.append(season)
    plt.figure(figsize=(10, 6))
    plt.title(feature, fontdict={'fontsize': 15})
    plt.boxplot(data, labels=labels)
```

```python
discrete_features = ['duration_ms', 'danceability', 'energy', 'loudness', 'acousticness', 'instrumentalness', 'liveness', 'valence']
```

```python
for discrete_feature in discrete_features:
    compare_feature_between_seasons(discrete_feature)
```
